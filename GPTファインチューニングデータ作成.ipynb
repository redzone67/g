##### 解説の特徴

````python
data = ''
with open('think.txt',mode='r',encoding='utf-8') as f:
    data += f.read()
with open('think2.txt',mode='r',encoding='utf-8') as f:
    data += f.read()
data =data.replace('\n','').replace('}{','}*{').split('*')
data = [s.replace('"予想"','\'予想\'').replace('"準備"','\'準備\'') for s in data]
data = [eval(s) for s in data]
````

````python
len(data)
````

````python
def analyze_cause_consequence(task, text, retry_delay=60, max_retries=3):
    retries = 0
    while retries < max_retries:
        try:
            # OpenAIのChatCompletion APIを使用してテキストを分析
            response = client.chat.completions.create(
                model="gpt-4-1106-preview",  # 適切なモデルを選択
                messages=[
                    {"role": "system", "content": "あなたは有能な編集者です。"},
                    {"role": "user", "content": task+text}
                ],
                temperature = 0
            )
            return response.choices[0].message.content
            
        except Exception as e:
            print(f"Error processing text: {text}. Retrying in {retry_delay} seconds... Error: {str(e)}")
            time.sleep(retry_delay)  # リトライ前に待機
            retries += 1
            
    return f"Failed to process text: {text} after {max_retries} retries."
````

````python
# レートリミットのパラメータ
requests_per_minute = 5000 # １秒あたりのリクエスト RPM = 5000
interval = 240/1000 #60 / requests_per_minute  # 各リクエスト間の間隔（秒）
````

````python
task ='''
次のデータは経済記事に対する解説で読者の評判が高いものです。
「注目を集める解説を書くためのノウハウ」を収集するため、
この解説から、その特徴を箇条書きで教えてください。
結果は次の形式に従ってください。
{"元の解説":,"解説の特徴":[...]}
余計な説明は不要です。No talk, just go.
#解説
'''
````

````python
results = []
start_time = datetime.now()
zero = datetime.now()

with ThreadPoolExecutor(max_workers=requests_per_minute) as executor:
    futures = []
    for s in data:
        text = s['解説']
        # 現在時刻と最後のリクエスト時刻の差が間隔より小さい場合、待機
        while (datetime.now() - start_time).total_seconds() < interval:
            time.sleep(0.01)  # 微小な待機時間
        start_time = datetime.now()
        futures.append(executor.submit(analyze_cause_consequence, task, text))
    
    for future in as_completed(futures):
        results.append(future.result())

print('終わりました:',datetime.now()-zero)
````

````python
point = []
for i in range(len(results)):
    point += eval(results[i])['解説の特徴']
````

````python
start_time = datetime.now()
model="text-embedding-3-small"
vector = []
for s in point:
    vector += [client.embeddings.create(input =[s], model=model).data[0].embedding]
print('終わりました:',datetime.now()-start_time)
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
#import matplotlib.pyplot as plt
# 階層クラスタリングを実行
Z = linkage(vector, 'ward')

# デンドログラムを描画
#plt.figure(figsize=(10, 7))
# クラスタ数
cln = 20

clusters = list(fcluster(Z ,t=cln, criterion='maxclust'))
cluster_label = {}
for i in range(len(clusters)):
    try:
        cluster_label[clusters[i]]+= point[i]+'\n'
    except:
        cluster_label[clusters[i]]=''
````

````python
for i in range(1,cln+1):
    print(i,'***************************************')
    print(cluster_label[i],'\n')
````
